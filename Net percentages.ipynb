```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
data = pd.read_csv("E:/UCLA/Bruin Sports Analytics/Tennis/UCLA_Spencer_Johnson_vs_TexasA&M_Tiago_Pires_1-17.csv")
data
```

```python
def is_odd(n):
  return n % 2 == 1
```

```python
volleys = pd.DataFrame(data.loc[(data["isVolley"] == 1) & (data["serverName"] == "Spencer Johnson") & (data["shotInRally"].apply(is_odd))])
num_volleys = len(volleys)
```

```python
approach = pd.DataFrame(data.loc[(data["isApproach"] == 1) & (data["serverName"] == "Spencer Johnson") & (data["shotInRally"].apply(is_odd))])
num_approach = len(approach)
```
