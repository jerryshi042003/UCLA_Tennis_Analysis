```python
import pandas as pd
import numpy as np
```

```python
def dashboard_stats(df, player_name):
    df["double_fault"] = 0
    first_serve_total = np.sum((df["firstServeIn"] == 1).astype(float))
    second_serve_total = np.sum((df["secondServeIn"] == 1).astype(float))
    df.loc[(df["firstServeIn"] == 0) & (df["secondServeIn"] == 0), "double_fault"] = 1
    double_fault_total = np.sum(df["double_fault"])
    ace_total = np.sum((df["isAce"] == 1).astype(float))
    first_serve_ad = np.sum(((df["firstServeIn"] == 1) & (df["serverName"] == player_name) & (df["side"] == "Ad")).astype(float))
    first_serve_deuce = np.sum(((df["firstServeIn"] == 1) & (df["serverName"] == player_name) & (df["side"] == "Deuce")).astype(float))
    second_serve_ad = np.sum(((df["secondServeIn"] == 1) & (df["serverName"] == player_name) & (df["side"] == "Ad")).astype(float))
    second_serve_deuce = np.sum(((df["secondServeIn"] == 1) & (df["serverName"] == player_name) & (df["side"] == "Deuce")).astype(float))
    double_fault_ad = np.sum(((df["double_fault"] == 1) & (df["serverName"] == player_name) & (df["side"] == "Ad")).astype(float))
    double_fault_deuce = np.sum(((df["double_fault"] == 1) & (df["serverName"] == player_name) & (df["side"] == "Deuce")).astype(float))
    serve_total = (first_serve_total+second_serve_total+double_fault_total)
    first_serve_percentage = (first_serve_total/serve_total)*100
    second_serve_percentage = (second_serve_total/serve_total)*100
    double_fault_percentage = (double_fault_total/serve_total)*100
    ad_total = (first_serve_ad.sum() + second_serve_ad.sum() + double_fault_ad.sum())
    deuce_total = (first_serve_deuce.sum() + second_serve_deuce.sum() + double_fault_deuce.sum())
    first_serve_ad_perc = np.round((first_serve_ad/ad_total)*100, 3)
    first_serve_deuce_perc = np.round((first_serve_deuce/deuce_total)*100, 3)
    second_serve_ad_perc = np.round((second_serve_ad/ad_total)*100, 3)
    second_serve_deuce_perc = np.round((second_serve_deuce/deuce_total)*100, 3)
    double_fault_ad_perc = np.round((double_fault_ad/ad_total)*100, 3)
    double_fault_deuce_perc = np.round((double_fault_deuce/deuce_total)*100, 3)
    match_title = str(df["clientTeam"][0]) + "-" + str(df["player1Name"][1]) + " " + "vs." + " " + str(df["opponentTeam"][0]) + "-" + str(df["player2Name"][1])
    match_date = df["Date"][0]
    dashboard_data = {
        "Date" : match_date,
        "Match" : match_title,
        "P Name": player_name,
        "Tot serves": serve_total,
        "Aces": ace_total,
        "1st Serve %": round(first_serve_percentage, 2),
        "2nd Serve %": round(second_serve_percentage, 2),
        "DF %": round(double_fault_percentage, 2),
        "1st serve-Ad": first_serve_ad,
        "1st serve-De": first_serve_deuce,
        "1st serve%-Ad": round(first_serve_ad_perc, 2),
        "1st serve%-De": round(first_serve_deuce_perc, 2),
        "2nd serve%-Ad": round(second_serve_ad_perc, 2),
        "2nd serve%-De": round(second_serve_deuce_perc, 2),
        "DF%-Ad": round(double_fault_ad_perc, 2),
        "DF%-De": round(double_fault_deuce_perc, 2)
    }
    dashboard = pd.DataFrame(dashboard_data, index = [0])
    dashboard.set_index("Date", inplace = True)
    return dashboard
```

```python
def summary_stats(player_name):
    df = main_dashboard.copy()
    df = df[df["P Name"] == player_name]
    summary = {   "Date" : np.NaN,
        "Match" : "Summary",
        "P Name": player_name,
        "Tot serves": df["Tot serves"].sum(),
        "Aces": df["Aces"].sum(),
        "1st Serve %": round(np.mean(df["1st Serve %"]), 2),
        "2nd Serve %": round(np.mean(df["2nd Serve %"]), 2),
        "DF %": round(np.mean(df["DF %"]), 2), 
        "1st serve-Ad": round(np.mean(df["1st serve-Ad"]), 2),
        "1st serve-De": round(np.mean(df["1st serve-De"]), 2),
        "1st serve%-Ad": round(np.mean(df["1st serve%-Ad"]), 2),
        "1st serve%-De": round(np.mean(df["1st serve%-De"]), 2),
        "2nd serve%-Ad": round(np.mean(df["2nd serve%-Ad"]), 2),
        "2nd serve%-De": round(np.mean(df["2nd serve%-De"]), 2),
        "DF%-Ad": round(np.mean(df["DF%-Ad"]), 2),
        "DF%-De": round(np.mean(df["DF%-De"]), 2)}
    df.loc[len(df)] = summary
    return df
```
