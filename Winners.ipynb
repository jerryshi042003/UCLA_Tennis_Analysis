```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
data = pd.read_excel("E:/UCLA/Bruin Sports Analytics/Tennis/Govind Nanda (UCLA) Raphael Perot (Texas A&M) Full Match.xlsx")
data
```

```python
def is_odd(n):
  return n % 2 == 1
```

```python
winners = pd.DataFrame(data.loc[(data["isWinner"] == 1) & (data["serverName"] == "Govind Nanda") & (data["shotInRally"].apply(is_odd))])
winners
```

```python
winner_percent = (len(winners)/len(data))*100
print(winner_percent)
```

```python
winner_fh = winners.loc[winners["shotType"] == "Forehand"]
winner_fh
```

```python
winner_bh = winners.loc[winners["shotType"] == "Backhand"]
winner_bh
```

```python
winner_fh_pct = (len(winner_fh)/len(winners))*100
print(winner_fh_pct)
```

```python
winner_bh_pct = (len(winner_bh)/len(winners))*100
print(winner_bh_pct)
```
